```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_87.csv',skiprows=4)
```

```python
selected_years = ["1960", "1970", "1980", "1990", "2000", "2010", "2020", "2023"]
```

```python
countries =  ["United States", "China", "India", "Brazil", "Germany", "Nigeria"]
df_filtered = df[df["Country Name"].isin(countries)][["Country Name"] + selected_years]
```

```python
df_melted = df_filtered.melt(id_vars=["Country Name"], var_name="Year", value_name="Population")
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Year", y="Population", hue="Country Name")
plt.title("Population Growth of Selected Countries Over Time")
plt.xticks(rotation=45)
plt.ylabel("Total Population")
plt.legend(title="Country")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["2023"].dropna(), bins=30, kde=True)
plt.title("Distribution of Population Across Countries (2023)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.xscale("log")  # Log scale to handle large population variations
plt.show()
```
